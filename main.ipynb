```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("teams.csv")
```

```python
data
```

```python
X = data[["athletes", "prev_medals"]].copy()
y = data[["medals"]].copy()
```

```python
X
```

```python
y
```

```python
X["constant"] = 1
```

```python
X
```

```python
X = X[["constant", "athletes", "prev_medals"]]
```

```python
X
```

```python
X_T = X.T
```

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y
```

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
y_pred = X @ B
```

```python
y_pred
```

```python
SSR = ((y-y_pred) ** 2).sum()
```

```python
SSR
```

```python
SST = ((y - y.mean()) ** 2).sum()
```

```python
SST
```

```python
r_sq = 1 - (SSR/SST)
```

```python
r_sq
```
